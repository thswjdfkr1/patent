```python
import pandas as pd

# Load the uploaded CSV file to examine its contents
file_path = '/content/drive/MyDrive/제발2.csv'
data = pd.read_csv(file_path)

# Display the first few rows of the dataframe
data.head()
```

```python
# Define keywords related to AI semiconductors
keywords = [
    "AI semiconductor", "neural network hardware", "machine learning chip",
    "deep learning accelerator", "인공지능 반도체", "뉴럴 네트워크 하드웨어",
    "머신러닝 칩", "딥러닝 가속기"
]

# Function to filter rows based on keywords
def filter_rows_by_keywords(df, keywords):
    pattern = '|'.join(keywords)
    filtered_df = df[
        df['발명의 명칭'].str.contains(pattern, case=False, na=False) |
        df['요약'].str.contains(pattern, case=False, na=False) |
        df['text_en'].str.contains(pattern, case=False, na=False)
    ]
    return filtered_df

# Apply the filtering function
filtered_data = filter_rows_by_keywords(data, keywords)

# Display the first few rows of the filtered dataframe
filtered_data.head()
```

```python
# Save the filtered data to a new CSV file
filtered_file_path = '//content/drive/MyDrive/aabbcc.csv'
filtered_data.to_csv(filtered_file_path, index=False)

filtered_file_path
```
