```python
import nltk
nltk.download("punkt")
from nltk import word_tokenize, bigrams
import pandas as pd
from nltk import ConditionalFreqDist
```

```python
df= pd.read_csv("/content/drive/MyDrive/특허/nn.csv", encoding='cp949')
```

```python
print(df)
```

```python
class1 = df['출원인_연도'].to_list()
class2 = df['메인 IPC'].to_list()
```

```python
docs = pd.DataFrame({'cs':class1, 'ipc':class2})
print(docs)
```

```python
docs_drop = docs.drop_duplicates(subset=['cs','ipc'])
print(docs_drop.shape)
```

```python
import networkx as nx
```

```python
graph = nx.from_pandas_edgelist(docs_drop, source = 'ipc', target = 'cs', create_using=nx.Graph())
```

```python
nx.write_gexf(graph,r"/content/drive/MyDrive/nn.gexf")
```
