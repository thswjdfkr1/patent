```python
from collections import Counter
import pandas as pd

data = pd.read_excel('/content/drive/MyDrive/asdasdas.xlsx').fillna("")
```

```python
# Extract the IPC codes
ipc_codes = data['전체 IPC']
```

```python
# Split the IPC codes on ',' and flatten the list
all_ipc_codes = [code for codes in ipc_codes for code in codes.split(',')]
# Count the frequency of each IPC code
ipc_counter = Counter(all_ipc_codes)

# Get the top 30 IPC codes
top_30_ipc = ipc_counter.most_common(30)
top_30_ipc_codes = [code for code, count in top_30_ipc]
```

```python
# Create a new column indicating if the IPC code is in the top 30
data['Is_Top_30'] = data['전체 IPC'].apply(lambda x: any(code in top_30_ipc_codes for code in x.split(',')))

# Save the updated dataframe to a new Excel file
output_file_path = '/content/drive/MyDrive/AAABCC.xlsx'
data.to_excel(output_file_path, index=False)

output_file_path
```
