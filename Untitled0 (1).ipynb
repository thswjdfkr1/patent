```python
import pandas as pd

# Load the Excel file
file_path = '/content/drive/MyDrive/리셋2.xlsx'
xlsx = pd.ExcelFile(file_path)

# Load the data from the relevant sheet
sheet2_df = pd.read_excel(xlsx, sheet_name='전체')

# Define keywords for filtering
keywords = ["반도체", "인공지능", "AI", "semiconductor", "artificial intelligence"]

# Filter the data based on keywords
filtered_df = sheet2_df[
    sheet2_df.apply(lambda row: any(keyword in str(row) for keyword in keywords), axis=1)
]

# Save the filtered data to a new Excel file
filtered_file_path = '/content/drive/MyDrive/리셋2_filtered_patents.xlsx'
filtered_df.to_excel(filtered_file_path, index=False)

print(f'Filtered data saved to {filtered_file_path}')
```
